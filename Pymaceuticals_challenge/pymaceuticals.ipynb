### Analysis

This study was requested by Pymaceuticals, Inc., from one of its senior data scientists.  The study examined a number of metrics to try to determine the relative efficacy of ten drug regimens that are intended to treat a very common form of skin cancer.  The purpose of this study was to compare results from pPymaceuticalsâ€™ drug, Capomulin, with the nine other treatment regimens.

The study began with 249 mice and ran for 45 days.  Weights of tumors were observed and measured at intervals throughout the study.  Some mice did not survive for the 45 days.

The executives at Pymaceuticals requested certain tables and figures for a technical report on the clinical study, and have also asked for a top-level summary of the study results.

The analysis included examing mean, median, variance, standard deviation, and standard err for the tumor sizes of mice being treated with each of the ten drug regimens.  The analysis also includeed bar, line, pie, and scatter plots to compare number of observations per drug regimen, the number of male and female mice, tumor size over time for an individual mouse, and tumor size to mouse weight.

All of the plots are included in this Jupyter notebook.


#### Observations

During data cleaning, data on one mouse was dropped before the analysis, because duplicate timepoint observations on that mouse made the results suspect.  

The study sample of 248 mice was approximately evenly divided between male and female mice, and amonog the ten different regimens under observation.  

Two of the drug regimens stand out clearly over the others as extending mouse life.  These are Capomulin, the drug manufactured by the company conducting the study, and Ramicane.  As can be seen in the bar charts, these two drugs produce approximately the same increase in the number of observations per mouse, with many mice in that study surviving to be observed at 45 days.

The other drug regimens all produce approximately 50 fewer observations, on average, with one exception, Propiva, that produces notably fewer observations by at least twenty on average than the 
other, less-successful drug regimens.   While two of the others, on this data, have a slightly larger number of observations per mouse, the differnences here are quite small.  Given the small sample size, however, the differnces might be much more noticeable in a much larger sample.

There is a strong correlation between mouse weight and tumor volume, with Pearson's r at .84.  This can be seen in the scatter plot comparing mouse weight and average tumor volume.

There also appears to be a strong correlation between reduction in tumor size and an increased number of observations (and thus longer mouse life).  The averge tumor sizes for both Capomulin and Ramicane were around 40 mm3, whereas the average tumor sizes in all of the other drug regmins ranged from approximately 52 to 54 mm3.  

The correlation is not very strong among the eight other drug regimens.  Mice given Propiva, which had notably the smallest number of observations, and accordingly the shortest survival times, had the smallest average tumor sizes of any of the eight drugs below the top two.  It may be that the differences in tumor weights for this group are too small to reach have any effect, because all of mice in the eight regimens had tumors that were very similar in size, differing on on average by less than 3 mm3.  The results could suggest that there is a cutoff in size in order to extend life expectancy, and all tumors over that cutoff mass will lead to death fairly rapidly.  This data does not indicate where between roughly 44 and 55 mm3 that cutoff might lie.


#### Take Aways and Next Steps

The test data size of 248 mice is quite a small sample, and divided between the ten different drug regimens under study, with a study period of 45 days, may be an inadequate representation of the effectiveness of any of these drug regimens.  Each mouse's individual responses are playing too large a role in the overall results.

Nonetheless, in this small study, several metrics are very notable as preliminarily suggestiong that the drug the study sponsor is manufacturing shows promise in extening mouse life.  The sponsor should also note that a second drug regimen appears to be approximately equally sucessful in extending life.  Because the cancer drug market is vast, this should not discourage Pymaceuticals from continuing to pursue development of its drug regimen.

None of the requested charts or metrics show whether there was a difference between tumor size or number of observations, e.g., more days of life, between male and female mice.  While any such difference might not carry over into humans, it would be something worth investigating.  If there is a difference in results between females and males, that could guide future studies, and, eventually,
if the drug regimen is successful, how the drug is prescribed and marketed.

Additinally, while the mice differed significantly in age (by more than one year, which is quite long in terms of mouse longevity), the analysis to date did not examine whether age may play a role in the observed outomes.  To some extent, the observations of mouse weight may be correlated with age, but so far that has not been established from the current analysis.  The existing data would allow such a comparison.

Because of the large variances and standard deviations seen in tumor size in all of the drug regimens, further studies should be conducted with much larger sample sizes.  Individual idiosyncratic responses may be responsible for variances of 20 mm3, with an average tumor size of 40 mm3.  These differences of almost half of the tumor size among different mice may reflect a successful drug regimen, but this data is too preliminary and the reasons for the results too unclear to determine why the average tumor size appears to have shrunk for some drug regimens.

# Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame

# start with the list of regimens

regimens = []
regimens = cleaned_mice["Drug Regimen"].unique()

#  Sort alphtabetically to match the group bys

regimens.sort()

regimens
```

```python
# create the base DataFrame

regimens_summary_df = pd.DataFrame({"Drug Regimen" : regimens})

# set the index and reorder for appending the additional group by columns

regimens_summary_df.set_index("Drug Regimen", drop=True, inplace=True)

regimens_summary_df.sort_index(ascending=True, inplace=True)
```

```python
# get a list of column names
col_names=cleaned_mice.columns

col_names
```

```python
# Group the mice by drug regimen to use in the statistics calculations

grouped_mice_df = cleaned_mice.groupby(["Drug Regimen"])

grouped_mice_df.head(10)
                                                    
```

```python
# Individually for each regimen, calculate the mean,
# median, variance, and standard deviation 
# of tumor volumes per regimen.
# Calculate the SEM of the standard deviation. 
# Append all new lists to the summary DataFrame


# Collect results for all regions
# Using numpy mean and median rather than Python native because calculation in C is faster

# arrays to hold output from all of the regions
means = []
medians =  []
variances = []
stds = []
stderrs = []
sems= []

# calculate the statistics for each region and append to lists
for regimen in regimens:
    
    regimen_vols = []
    regimen_mean = 0
    regimen_median = 0
    regimen_variance = 0
    regimen_std = 0
    regimen_stderr = 0
    
    regimen_df = pd.DataFrame(cleaned_mice.loc[cleaned_mice["Drug Regimen"] == regimen, :])
    regimen_vols = regimen_df["Tumor Volume (mm3)"]
    
    regimen_mean = np.mean(regimen_vols)
    regimen_median = np.median(regimen_vols)
    
    # np.std and np.var calculate the standard deviation and variance for 
    # the entire population,using N.  Our data is for the entire population, 
    # so this would be the correct way to do it.  The Python "describe()" method  
    # caculates for a sample of the population, using N-1, as do the DataFrame
    # "var" and "sem" methods.  To make the results match, use
    # one degree of freedom (ddoff = 1), in numpy.  The requirement of the 
    # assignment is to match, even though the instructions say that we have 
    # all of the data used in the study.  The study data, however, could be 
    # viewed as a sample of the entire population of mice with tumors.
    
    regimen_variance = np.var(regimen_vols, ddof=1)
    regimen_std = np.std(regimen_vols, ddof=1)
    
    # The default value of sciypy sem() is ddof = 1.
    # This is different than the default for other packages
    # such as DataFrame sem() method, so by using
    # the defaults here the results will match.
    
    regimen_stderr = st.sem(regimen_vols)
    
    means.append(regimen_mean)
    medians.append(regimen_median)
    variances.append(regimen_variance)
    stds.append(regimen_std)
    sems.append(regimen_stderr)

# add the results for all regimen to the summary DataFrame

regimens_summary_df["Mean Tumor Vol"] = means
regimens_summary_df["Median Tumor Vol"] = medians
regimens_summary_df["Variance"] = variances
regimens_summary_df["STD"] = stds
regimens_summary_df["SEM"] = sems
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

grouped_mice = cleaned_mice.groupby(["Mouse ID"])

max_timepoints_df = grouped_mice["Timepoint"].max()
max_timepoints_df=max_timepoints_df.reset_index()

# merge with original cleaned mice to get volume for that last timepoint

max_mice_df = max_timepoints_df.merge(cleaned_mice, how="left", on=["Mouse ID", "Timepoint"])

max_mice_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)

tot_treatment_vols = []
outlier_results = ""

# Loop through each drug in the treatment list, locating the rows in the merged DataFrame that correspond to each treatment. Append the resulting final tumor volumes for each drug to the empty list.

# get the max tumor vol for each treatment and append to total olume list

for treatment in treatments:
    
    outlier_results += (f"For drug regiment {treatment}:\n\n")
    
    treatment_vols_df = pd.DataFrame(max_mice_df.loc[max_mice_df["Drug Regimen"] == treatment, :])
        
    # get all tumor vol rows for this treatment
    
    single_treatment_vols_df = treatment_vols_df["Tumor Volume (mm3)"]
    
    # add the values from this treatment regimen to the totals array
    tot_treatment_vols.append((single_treatment_vols_df.values))
    
    #--------------------------------------------
    # Determine outliers by using the upper and lower bounds .
    #---------------------------------------------

    quart1 = single_treatment_vols_df.describe().iloc[4]
    quart3 = single_treatment_vols_df.describe().iloc[6]
    
    # Calculate the IQR and quantitatively determine if there are any 
    # potential outliers
    
    IQR = quart3 -quart1
    
    outlier_below = quart1 - (1.5*IQR)
    outlier_above = quart3 + (1.5 * IQR)
    
    outlier_results += (f"The first quartile is: {quart1:2.4f}\n")
    outlier_results += (f"The third quartile is: {quart3:2.4f}\n")
    outlier_results += (f"The IQR is: {IQR:2.4f}\n")
    outlier_results += (f"Values below {outlier_below:2.4f} may be outliers\n")
    outlier_results += (f"Values above {outlier_above:2.4f} may be outliers\n")
    outlier_results += (f"\n\n")


# print results for all treatments
print(outlier_results)
 
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

# setoutlier properties

box_flierprops = {'marker': 'o', 'markerfacecolor': 'red', 'markeredgecolor': 'red', 'markersize': 8}

# Create box plot
plt.boxplot(tot_treatment_vols, showmeans=True, labels=treatments, flierprops=box_flierprops)
plt.grid()

# configure the plot

treatment_title = "Tumor Volumes for Selected Treatment Regimens"

plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Vol (mm3)')
plt.title(treatment_title, fontdict={'fontsize': 13})
plt.figsize=(8,6)

            
#display           
plt.show()
```

## Correlation and Regression

### Correlation

```python
# get average tumor volumes to compare to mouse weights for the selected regimen

regimen_mice_grouped = (cleaned_mice.loc[cleaned_mice["Drug Regimen"] == sample_drug_regimen, :]).groupby(["Mouse ID"])


regimen_avg_vols=regimen_mice_grouped["Tumor Volume (mm3)"].mean()
regimen_avg_weights=regimen_mice_grouped["Weight (g)"].mean()


# create the dataframe to hold the mice weights and tumor volumes

regimen_mouse_vol_weights = pd.DataFrame({'Mouse Weight' : regimen_avg_weights,
                                       'Tumor Vol' : regimen_avg_vols
                              })
```

```python
# Generate a scatter plot of mouse weight vs. average tumor volume for the entire Capomulin regimen

# setup formatting

regimen_title = (f"Mouse Weight vs. Average Tumor Volume for the {sample_drug_regimen} Regimen")

regimen_color="blue"

x_axis = np.arange(len(regimen_mouse_weights))


# create the scatter plot

fig = regimen_mouse_vol_weights.plot.scatter(x="Mouse Weight",
                                             y="Tumor Vol",
                                             s=15,
                                             c=regimen_color,
                                             title=regimen_title,
                                             figsize=(8,6),
                                             xlabel= "Mouse Weight (g)",
                                             ylabel="Tumor Volume (mm3)"
                                            )

# display
plt.show()
```

```python
# Calculate the correlation coefficient between mouse weight and average tumor volume

weights=regimen_mouse_vol_weights['Mouse Weight'].to_numpy()
vols=regimen_mouse_vol_weights['Tumor Vol'].to_numpy()

pearsons_r = st.pearsonr(weights, vols)

weight_corr = pearsons_r[0]

print(f"The correlation betweeen mouse weight and tumour volume is {round(weight_corr,2)}.")
```

### Linear Regression

```python
# Perform a linear regression on mouse weogjt versus average tumor volume for study regimen

reg_slope, reg_int, reg_r, reg_p, reg_std_err = st.linregress(weights, vols)

reg_slope                                                                                  
```

```python
# Create equation of line to calculate predicted tumor weight

reg_fit = reg_slope * weights + reg_int
```

```python
# Plot the linear model on top of scatter plot 

# using matplotlib this time

plt.scatter(weights, vols, color=regimen_color)

plt.plot(weights, reg_fit,"--", color="lightcoral")

plt.xticks(weights)
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Vol (mm3)')
plt.title(regimen_title, fontdict={'fontsize': 13})
plt.figsize=(8,6)

# display
plt.show()
```
