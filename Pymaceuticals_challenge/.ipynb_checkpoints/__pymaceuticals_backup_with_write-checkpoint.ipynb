```python
# Dependencies and Setup

import string
from pathlib import Path
import random

import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame

raw_combined_results = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview

raw_combined_results.head()
```

```python
# Checking the number of mice.
# Per Stack Overflow, use shape() rather than len() to handle
# null or NAN values that len does not count.

num_mice = (raw_combined_results["Mouse ID"].unique()).shape[0]

num_mice
```

```python
all_mice_observations = len(raw_combined_results)

all_mice_observations
```

```python
# The data should be uniquely identified by Mouse ID and Timepoint

# Getting the duplicates methods and dropping the duplicates are based on a combination of 
# information in chatGPT and w3chools.com/python, and the python documentation.

# first identify duplicate rows based on Mouse ID and Timepoint

duplicate_time_rows = raw_combined_results.duplicated(subset=["Mouse ID", "Timepoint"], keep = False)
 

# subset the dataframe to show only the duplicate rows of mouse id/time pairs

duplicates_df = raw_combined_results[duplicate_time_rows]

duplicates_df.head(15)
```

```python
# Get the duplicate mouse ID from the above duplicate pairs

duplicate_mouse  = duplicates_df["Mouse ID"].unique()

duplicate_mouse_id = duplicate_mouse[0]

duplicate_mouse_id
```

```python
# Optional:  Display all rows for the duplicate mouse ID

duplicate_rows = raw_combined_results[(raw_combined_results["Mouse ID"] == duplicate_mouse_id)]

duplicate_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

# Remove the duplicates from the combined results
# Don't keep the first instance of time points for the mouse, just drop all data for that mouse

cleaned_mice = raw_combined_results[(raw_combined_results["Mouse ID"] != duplicate_mouse_id)]

# save output for testing
combined_results_path = "output/cleaned_mice.csv"

# Read the mouse data and the study results
cleaned_mice.to_csv(combined_results_path)

cleaned_mice.tail(10)
```

# Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame

# start with the list of regimens

regimens = []
regimens = cleaned_mice["Drug Regimen"].unique()

#  Sort alphtabetically to match the group bys

regimens.sort()

regimens
```

```python
# create the base DataFrame

regimens_summary_df = pd.DataFrame({"Drug Regimen" : regimens})

# set the index and reorder for appending the additional group by columns

regimens_summary_df.set_index("Drug Regimen", drop=True, inplace=True)

regimens_summary_df.sort_index(ascending=True, inplace=True)
```

```python
# get a list of column names
col_names=cleaned_mice.columns

col_names
```

```python
# Group the mice by drug regimen to use in the statistics calculations

grouped_mice_df = cleaned_mice.groupby(["Drug Regimen"])

grouped_mice_df.head(10)
                                                    
```

```python
# Individually for each regimen, calculate the mean,
# median, variance, and standard deviation 
# of tumor volumes per regimen.
# Calculate the SEM of the standard deviation. 
# Append all new lists to the summary DataFrame


# Collect results for all regions
# Using numpy mean and median rather than Python native because calculation in C is faster

# arrays to hold output from all of the regions
means = []
medians =  []
variances = []
stds = []
stderrs = []
sems= []

# calculate the statistics for each region and append to lists
for regimen in regimens:
    
    regimen_vols = []
    regimen_mean = 0
    regimen_median = 0
    regimen_variance = 0
    regimen_std = 0
    regimen_stderr = 0
    
    regimen_df = pd.DataFrame(cleaned_mice.loc[cleaned_mice["Drug Regimen"] == regimen, :])
    regimen_vols = regimen_df["Tumor Volume (mm3)"]
    
    regimen_mean = np.mean(regimen_vols)
    regimen_median = np.median(regimen_vols)
    
    # np.std and np.var calculate the standard deviation and variance for 
    # the entire population,using N.  Our data is for the entire population, 
    # so this would be the correct way to do it.  The Python "describe()" method  
    # caculates for a sample of the population, using N-1, as do the DataFrame
    # "var" and "sem" methods.  To make the results match, use
    # one degree of freedom (ddoff = 1), in numpy.  The requirement of the 
    # assignment is to match, even though the instructions say that we have 
    # all of the data used in the study.  The study data, however, could be 
    # viewed as a sample of the entire population of mice with tumors.
    
    regimen_variance = np.var(regimen_vols, ddof=1)
    regimen_std = np.std(regimen_vols, ddof=1)
    
    # The default value of sciypy sem() is ddof = 1.
    # This is different than the default for other packages
    # such as DataFrame sem() method, so by using
    # the defaults here the results will match.
    
    regimen_stderr = st.sem(regimen_vols)
    
    means.append(regimen_mean)
    medians.append(regimen_median)
    variances.append(regimen_variance)
    stds.append(regimen_std)
    sems.append(regimen_stderr)

# add the results for all regimen to the summary DataFrame

regimens_summary_df["Mean Tumor Vol"] = means
regimens_summary_df["Median Tumor Vol"] = medians
regimens_summary_df["Variance"] = variances
regimens_summary_df["STD"] = stds
regimens_summary_df["SEM"] = sems
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.

# get the totals of the mouse/timepoin pairs per region

total_rows = grouped_mice_df["Tumor Volume (mm3)"].count()

# order the results so that the regimen with the longest-living mice appear first

total_rows.sort_values(inplace = True, ascending=False)
ordered_regimens = total_rows.index

# set the locations for the ticks per region

x_axis = np.arange(len(ordered_regimens))

# create the bar plot

fig = total_rows.plot.bar(color="blue", figsize=(8,6),
                                   title="Number of Mouse/Timepoints Observations Per Regimen",
                                   xlabel= "Drug Regimen",
                                   ylabel="Mouse/Timepoint Observations",
                                   alpha=0.75 )
# set the size
fig.title.set_size(14)

# display
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot

# The figsize has to be set first before the plot call 
plt.figure(figsize=(8, 6))

# create the bar plot, set the size of the bars, dim color tone slightly

fig = plt.bar(x_axis, total_rows, facecolor="blue", alpha=0.75, align="center", width=.5)  


# Configure the appearance of the chart

plt.title("Number of Mouse/Timepoints Observations Per Regimen", fontdict={'fontsize': 14})
plt.title.loc=('center')

# setup the ticks for the regions
x_axis = np.arange(len(ordered_regimens))
        
tick_locations = []
for x in x_axis:
    tick_locations.append(x)

plt.xticks(tick_locations, ordered_regimens, rotation=90)

# set the lables for the axes
plt.xlabel("Drug Regimen")
plt.ylabel("Mouse/Timepoint Observations")
           
# display
plt.show()
```
